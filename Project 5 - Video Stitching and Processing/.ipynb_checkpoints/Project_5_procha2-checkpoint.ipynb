### Part I: Stitch two key frames 

#### This involves:
1. compute homography H between two frames; 
2. project each frame onto the same surface;
3. blend the surfaces.

Check that your homography is correct by plotting four points that form a square in frame 270 and their projections in each image.

```python
def score_projection(pt1, pt2):
  '''
  Score corresponding to the number of inliers for RANSAC
  Input: pt1 and pt2 are 2xN arrays of N points such that pt1[:, i] and pt2[:,i] should be close in Euclidean distance if they are inliers
  Outputs: score (scalar count of inliers) and inliers (1xN logical array)
  '''

  # TO DO
  # pt1 = Xb[:2,:]/Xb[2,:]
  # pt2 = Xb_[:2,:]/Xb_[2,:]
    
  u = pt2[0] - pt1[0]
  v = pt2[1] - pt1[1]

  inliers = np.sqrt(u**2 + v**2) < 3
  score = sum(inliers)

  return score, inliers

def auto_homography(Ia,Ib, homography_func=None,normalization_func=None):
    '''
    Computes a homography that maps points from Ia to Ib

    Input: Ia and Ib are images
    Output: H is the homography

    '''
    if Ia.dtype == 'float32' and Ib.dtype == 'float32':
        Ia = (Ia*255).astype(np.uint8)
        Ib = (Ib*255).astype(np.uint8)
    
    Ia_gray = cv2.cvtColor(Ia,cv2.COLOR_BGR2GRAY)
    Ib_gray = cv2.cvtColor(Ib,cv2.COLOR_BGR2GRAY)

    # Initiate SIFT detector
    sift = cv2.xfeatures2d.SIFT_create()
    
    # find the keypoints and descriptors with SIFT
    kp_a, des_a = sift.detectAndCompute(Ia_gray,None)
    kp_b, des_b = sift.detectAndCompute(Ib_gray,None)    
    
    # BFMatcher with default params
    bf = cv2.BFMatcher()
    matches = bf.knnMatch(des_a,des_b, k=2)

    # Apply ratio test
    good = []
    for m,n in matches:
        if m.distance < 0.75*n.distance:
            good.append(m)
   
    numMatches = int(len(good))

    matches = good

    # Xa and Xb are 3xN matrices that contain homogeneous coordinates for the N
    # matching points for each image
    Xa = np.ones((3,numMatches))
    Xb = np.ones((3,numMatches))
    
    for idx, match_i in enumerate(matches):
        Xa[:,idx][0:2] = kp_a[match_i.queryIdx].pt
        Xb[:,idx][0:2] = kp_b[match_i.trainIdx].pt

    ## RANSAC
    niter = 1000
    best_score = 0
    n_to_sample = 4 #???? # Put the correct number of points here

    for t in range(niter):
        # estimate homography
        subset = np.random.choice(numMatches, n_to_sample, replace=False)
        pts1 = Xa[:,subset]
        pts2 = Xb[:,subset]
        
        H_t = homography_func(pts1, pts2, normalization_func) # edit helper code below (computeHomography)

        # score homography
        Xb_ = np.dot(H_t, Xa) # project points from first image to second using H        
        score_t, inliers_t = score_projection(Xb[:2,:]/Xb[2,:], Xb_[:2,:]/Xb_[2,:])
        
        if score_t > best_score:
            best_score = score_t
            H = H_t
            in_idx = inliers_t
    
    print('best score: {:02f}'.format(best_score))

    # Optionally, you may want to re-estimate H based on inliers

    return H
```

```python
def computeHomography(pts1, pts2,normalization_func=None):
    '''
    Compute homography that maps from pts1 to pts2 using SVD. Normalization is optional.
     
    Input: pts1 and pts2 are 3xN matrices for N points in homogeneous
    coordinates. 
    
    Output: H is a 3x3 matrix, such that pts2~=H*pts1
    '''
    # TO DO
    totalpoints = pts1.shape[1]
    
    A = np.zeros((2*totalpoints, 9))
    for i in range(totalpoints):
        A[2*i,:]   = [-pts1[0,i],-pts1[1,i],-1,0,0,0,pts1[0,i]*pts2[0,i],pts1[1,i]*pts2[0,i],pts2[0,i]]
        A[2*i+1,:] = [0,0,0,-pts1[0,i],-pts1[1,i],-1,pts1[0,i]*pts2[1,i],pts1[1,i]*pts2[1,i],pts2[1,i]]
    
    H,S,V = svd(A)
    h = V[-1,:]
    
    H = np.reshape(h,(3,3))
    
    return H
```

```python
# images location
im1 = './images/input/frames/f0270.jpg'
im2 = './images/input/frames/f0450.jpg'

# Load an color image in grayscale
im1 = cv2.imread(im1)
im2 = cv2.imread(im2)

H = auto_homography(im1,im2, computeHomography)
print(H/H.max()) 

# plot the frames here
box_pts = np.array([[300, 400, 400, 300, 300], [100, 100, 200, 200, 100], [1, 1, 1, 1, 1]])
plt.figure()
plt.imshow(im1[:,:,[2,1,0]])
plt.plot(box_pts[0,:], box_pts[1, :], 'r-')

# TO DO: project points into im2 and display the projected lines on im2
```

### Part II: Panorama using five key frames

Produce a panorama by mapping five key frames [90, 270, 450, 630, 810] onto the same reference frame 450.  


```python
key_frames_idx = np.array([90, 270, 450, 630, 810])-1

frames = np.zeros((len(key_frames_idx), im1.shape[0], im1.shape[1], im1.shape[2]),dtype='uint8')
for n in range(len(key_frames_idx)):
  frames[n] = cv2.imread("./images/input/frames/f0{num}.jpg".format(num=str(key_frames_idx[n]+1).zfill(3)))

# TO DO solution
H_270_450 = auto_homography(frames[1],frames[2], computeHomography)
H_630_450 = auto_homography(frames[3],frames[2], computeHomography)

H_90_270 = auto_homography(frames[0],frames[1], computeHomography)
H_810_630 = auto_homography(frames[4],frames[3], computeHomography)

H_90_450 = np.dot(H_90_270,H_270_450)
H_810_450 = np.dot(H_810_630,H_630_450)

panorama_w = 1800 # giving a little buffer beyond the original recommendation for better results
panorama_h = 780

H_t = [[1,0,600],[0,1,50],[0,0,1]]
H_t2 = [[1,0,600],[0,1,50]]

img_90 = cv2.warpPerspective(frames[0], np.dot(H_t,H_90_450), (panorama_w, panorama_h))
img_270 = cv2.warpPerspective(frames[1], np.dot(H_t,H_270_450), (panorama_w, panorama_h))
img_450 = cv2.warpAffine(frames[2], np.float32(H_t2), (panorama_w, panorama_h))
img_630 = cv2.warpPerspective(frames[3], np.dot(H_t,H_630_450), (panorama_w, panorama_h))
img_810 = cv2.warpPerspective(frames[4], np.dot(H_t,H_810_450), (panorama_w, panorama_h))

img_90 = np.float32(img_90/255.0)
img_270 = np.float32(img_270/255.0)
img_450 = np.float32(img_450/255.0)
img_630 = np.float32(img_630/255.0)
img_810 = np.float32(img_810/255.0)

canvas = np.zeros((img_90.shape))

#Build the final image
for i in range(canvas.shape[0]):
    for j in range(canvas.shape[1]):
        if img_810[i,j].any():
            canvas[i,j] = img_810[i,j]
        elif img_630[i,j].any():
            canvas[i,j] = img_630[i,j]
        elif img_450[i,j].any():
            canvas[i,j] = img_450[i,j]
        elif img_270[i,j].any():
            canvas[i,j] = img_270[i,j]
        elif img_90[i,j].any():
            canvas[i,j] = img_90[i,j]
            
fig, axes = plt.subplots(1, 1, figsize=(15,15))
axes.imshow(canvas) 
axes.set_title('Panorama with five framesmspain', fontsize=20) 
fig.tight_layout()
```

### Part 3: Map the video to the reference plane

Project each frame onto the reference frame (using same size panorama) to create a video that shows the portion of the panorama revealed by each frame

```python
# read all the images
import os 
dir_frames = 'images/input/frames'
filenames = []
filesinfo = os.scandir(dir_frames)

filenames = [f.path for f in filesinfo if f.name.endswith(".jpg")]
filenames.sort(key=lambda f: int(''.join(filter(str.isdigit, f))))

frameCount = len(filenames)
frameHeight, frameWidth, frameChannels = cv2.imread(filenames[0]).shape
frames = np.zeros((frameCount, frameHeight, frameWidth, frameChannels),dtype='uint8')

for i, file_i in enumerate(filenames):
  frames[i] = cv2.imread(file_i)
```

```python
# TO DO part 3 solution

total_images = 900
p3_path = "images/output/part3/"

HArray = np.zeros((total_images,3, 3),dtype=np.float32)
image_array = np.zeros((total_images,panorama_h, panorama_w, 3), dtype='uint8')

start_time = time.time()
print ("Starting computation at " + time.strftime("%H:%M:%S"))

#cleanup the folder before I start
for f in os.listdir(p3_path):
    os.remove(os.path.join(p3_path, f))
    
frame90 = cv2.imread('images/input/frames/f0090.jpg')
frame270 = cv2.imread('images/input/frames/f0270.jpg')
frame450 = cv2.imread('images/input/frames/f0450.jpg')
frame630 = cv2.imread('images/input/frames/f0630.jpg')
frame810 = cv2.imread('images/input/frames/f0810.jpg')

junction270_450 = auto_homography(frame270,frame450, computeHomography)
junction630_450 = auto_homography(frame630,frame450, computeHomography)
junction90_270 = auto_homography(frame90,frame270, computeHomography)
junction810_630 = auto_homography(frame810,frame630, computeHomography)
junction90_450 = np.dot(junction90_270,junction270_450)
junction810_450 = np.dot(junction810_630,junction630_450)

H_t = [[1,0,600],[0,1,50],[0,0,1]] # X and Y positions
for i, file_index in enumerate(filenames):
    print("Processing image " + str(i))
    frame = cv2.imread(file_index)
    
    if i < 90:
        homography = np.dot(auto_homography(frame,frame90, computeHomography),junction90_450)
        
    elif (i < 270) and (i >= 90):
        homography = np.dot(auto_homography(frame, frame270, computeHomography),junction270_450)
        
    elif (i >= 270) and (i <= 630):
        homography = auto_homography(frame,frame450, computeHomography)
        
    elif (i > 630) and (i <= 810):
        homography = np.dot(auto_homography(frame, frame630, computeHomography),junction630_450)
        
    elif i > 810:
        homography = np.dot(auto_homography(frame, frame810, computeHomography),junction810_450)
        
    newimage = cv2.warpPerspective(frame, np.dot(H_t,homography), (panorama_w, panorama_h))
    cv2.imwrite(p3_path + '{:03d}.jpg'.format(i), newimage)
    HArray[i,:,:] = np.dot(H_t,homography)
    image_array[i,:,:,:] = newimage

print ("Completed computation. Will now create Part 3 video.")

utils.imageFolder2mpeg(p3_path, output_path='part3_video.mpeg', fps=30)

end_time = time.time()
print ("Elapsed time (MM:SS): " + time.strftime("%M:%S", time.gmtime((end_time - start_time))))
```
